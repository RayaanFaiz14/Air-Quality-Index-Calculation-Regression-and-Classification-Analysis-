# Calculating Air Quality Index Based on Formula

The air quality index is a linear function of pollutant concentration.

<h3> Link to theory and calculation of AQI in India <h3>
   <h5> <a>https://www.pranaair.com/blog/what-is-air-quality-index-aqi-and-its-calculation/<a> <h5>

    

```python
def cal_SOi(so2):
    si=0
    if(so2<=40):
        si=so2*(50/40)
    elif(so2>40 and so2<=80):
        si=50+(so2-40)*(50/40)
    elif(so2>80 and so2<=380):
        si=100+(so2-80)*(100/300)
    elif(so2>380 and so2<=800):
        si=200+(so2-380)*(100/420)
    elif(so2>800 and so2<=1600):
        si=300+(so2-800)*(100/800)
    elif(so2>1600):
        si=400+(so2-1600)*(100/800)
    return si
df['SOi']=df['so2'].apply(cal_SOi)
data=df[['so2','SOi']]
data.head()
```

<h3> Function to calcuate NO2 individual pollutant index (ni)

```python
def cal_NOi(no2):
    ni=0
    if(no2<=40):
        ni=no2*(50/40)
    elif(no2>40 and no2<=80):
        ni=50+(no2-40)*(50/40)
    elif(no2>80 and no2<=180):
        ni=100+(no2-80)*(100/100)
    elif(no2>180 and no2<=280):
        ni=200+(no2-180)*(100/100)
    elif(no2>280 and no2<=400):
        ni=300+(no2-280)*(100/120)
    else:
        ni=400+(no2-400)*(100/120)
    return ni
df['NOi']=df['no2'].apply(cal_NOi)
data=df[['no2','NOi']]
data.head()
```

<h3> Function to calculate RSPM individual pollutant index (rpi) <h3>

```python
def cal_RSPMi(rspm):
    rpi=0
    if(rpi<=30):
        rpi=rpi*(50/30)
    elif(rpi>30 and rpi<=60):
        rpi=50+(rpi-30)*(50/30)
    elif(rpi>60 and rpi<=90):
        rpi=100+(rpi-60)*(100/30)
    elif(rpi>90 and rpi<=120):
        rpi=200+(rpi-90)*(100/30)
    elif(rpi>120 and rpi<=250):
        rpi=300+(rpi-120)*(100/130)
    else:
        rpi=400+(rpi-250)*(100/130)
    return rpi
df['RSPMi']=df['rspm'].apply(cal_RSPMi)
data=df[['rspm','RSPMi']]
data.head()
```

<h3> Function to calculate SPM individual pollutant index (spi) <h3>

```python
def cal_SPMi(spm):
    spi=0
    if(spm<=50):
        spi=spm*(50/50)
    elif(spm>50 and spm<=100):
        spi=50+(spm-50)*(50/50)
    elif(spm>100 and spm<=250):
        spi=100+(spm-100)*(100/150)
    elif(spm>250 and spm<=350):
        spi=200+(spm-250)*(100/100)
    elif(spm>350 and spm<=430):
        spi=300+(spm-350)*(100/80)
    else:
        spi=400+(spm-430)*(100/430)
    return spi
df['SPMi']=df['spm'].apply(cal_SPMi)
data=df[['spm','SPMi']]
data.head()
```

<h2> Function to calculate the AQI of every data value <h2>

```python
def cal_aqi(si,ni,rspmi,spmi):
    aqi=0
    if(si>ni and si>rspmi and si>spmi):
        aqi=si
    if(ni>si and ni>rspmi and ni>spmi):
        aqi=ni
    if(rspmi>ni and rspmi>si and rspmi>spmi):
        aqi=rspmi
    if(spmi>ni and spmi>rspmi and spmi>si):
        aqi=spmi
    return aqi

df['AQI']=df.apply(lambda x:cal_aqi(x['SOi'],x['NOi'],x['RSPMi'],x['SPMi']), axis=1)
data=df[['state','SOi','NOi','RSPMi','SPMi','AQI']]
data.head()
```

<h2> Air Category Based on AQI Score <h2>

<img src='https://w.ndtvimg.com/sites/3/2019/12/18122812/air_pollution_standards_cpcb.png'>

```python
def AQI_Range(x):
    if x<=50:
        return "Good"
    elif x>50 and x<=100:
        return "Satisfactory"
    elif x>100 and x<=200:
        return "Moderate"
    elif x>200 and x<=300:
        return "Poor"
    elif x>300 and x<=400:
        return "Severe"
    elif x>400:
        return "Hazardous"

df['AQI_Range']=df['AQI'].apply(AQI_Range)
df.head()
```
